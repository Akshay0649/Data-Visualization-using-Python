```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

**IMPORT DATASET**

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# @title Total Bill vs. Tip Amount by Gender

import matplotlib.pyplot as plt

plt.scatter(df[df['sex'] == 'Male']['total_bill'], df[df['sex'] == 'Male']['tip'], label='Male', alpha=0.7)
plt.scatter(df[df['sex'] == 'Female']['total_bill'], df[df['sex'] == 'Female']['tip'], label='Female', alpha=0.7)
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Total Bill vs. Tip Amount by Gender')
_ = plt.legend()
```

```python
plt.scatter(x= 'tip', y = 'total_bill', data = df, c ='g', s =40)

#Plotting the siz of coloumn
plt.scatter(x= 'tip', y = 'size', data = df, c ='y', marker = '+')

#Legends
plt.legend(['Total_bill', 'size'],title = 'Scatterplot')

#Labelling
plt.xlabel('Tip')
plt.ylabel('Total_bill and Size')
plt.title('Tip vs Total Bill && Size')
```

```python
#Plotting in Seaborn
sns.lmplot(x='tip', y='total_bill', data=df, fit_reg=True)
```

```python
# prompt: Can you explain the above chart

# The first plot you created is a scatter plot showing the relationship between 'Total Bill' and 'Tip Amount', with the points colored by 'Gender'.
# It helps visualize if there's any difference in tipping behavior between males and females.

# The second plot is another scatter plot, this time showing 'Tip' on the x-axis and both 'Total Bill' and 'Size' on the y-axis.
# The 'Total Bill' is represented by green circles and 'Size' by yellow plus markers. This plot might be a bit confusing as it tries to represent two relationships on the same axes.

# The third plot is created using Seaborn's `lmplot` and shows the relationship between 'Tip' and 'Total Bill' with a fitted regression line.
# This helps visualize the trend and potential linear relationship between these two variables.

# To improve the visualizations, you could consider:

# - Creating separate plots for 'Tip vs Total Bill' and 'Tip vs Size' to avoid overlapping data.
# - Using different markers or colors for the second plot to better distinguish between 'Total Bill' and 'Size'.
# - Adding a title and axis labels to all plots for clarity.

# Overall, these plots provide insights into the relationships between different variables in the 'tips' dataset,
# but they could be refined for better readability and interpretation.
```

```python
# prompt: Plot an Histogram

# Histogram of Total Bill
plt.hist(df['total_bill'], bins=10, edgecolor='black')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill')
plt.show()
```

```python
#seaborn Libaray Histogram
plt.figure()
sns.histplot(df.tip, kde = True)
```

```python
# prompt: What is kde in above histplot chart

# kde stands for Kernel Density Estimation.
# In the context of the seaborn histplot, setting kde=True adds a smoothed curve to the histogram,
# representing the estimated probability density function of the data.
# This can help visualize the underlying distribution of the data more smoothly than just the bars of the histogram.
```

```python
df.groupby(['time', 'day'])['tip'].agg(['mean'])
```

```python
#Bar plot
df.groupby(['time', 'day'])['tip'].mean().plot(kind = 'barh')
```

```python
#Stacked Bar Chart
plt.figure()
df.groupby(['time', 'day'])['tip'].agg(['mean', 'std']).plot(kind ='barh', stacked = True)
```

```python
# prompt: Explain this code
# #Stacked Bar Chart
# plt.figure()
# df.groupby(['time', 'day'])['tip'].agg(['mean', 'std']).plot(kind ='barh', stacked = True)

# This code creates a stacked horizontal bar chart using the 'tips' dataset in pandas and matplotlib.

# 1. Grouping and Aggregation:
#    - df.groupby(['time', 'day'])['tip'] groups the data by the 'time' and 'day' columns and selects the 'tip' column.
#    - .agg(['mean', 'std']) calculates the mean and standard deviation of the 'tip' values for each group.

# 2. Plotting:
#    - .plot(kind='barh', stacked=True) creates a horizontal bar chart (barh) and stacks the bars representing the mean and standard deviation on top of each other for each group.
#    - plt.figure() creates a new figure for the plot.

# Interpretation:
# - The resulting chart shows the average tip amount ('mean') for different combinations of 'time' and 'day', with the height of each bar segment representing the mean tip.
# - The stacked portion of each bar represents the standard deviation ('std') of the tips for that group, giving an idea of the variability in tip amounts.
```
